```python
#Bai 4
h=int(input("Nhập chiều cao tam giác:"))
for i in range(1,h+1):
    space = " " * (h-i)
    if i==1:
        print(space + "*")
    elif i==h:
        print("*" * (2*h-1))
    else:
        khoang_trong = " " * (2*i-3)
        print(space + "*" + khoang_trong + "*")

```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n=int(input())
arr= []
c=0
l=0
while n>0:
    k=n%10
    arr.append(k)
    n=(n-k)/10

for i in arr:
    if i%2==0:
        c+=1
    else:
        l+=1
print("So so chan la:",c,"So so le la:",l)




```

```python
n=input()
c=0
l=0
for s in n:
    so=int(s)
    if so%2==0:
        c+=1
    else:
        l+=1
print("So so chan la:", c)
print("So so le la:", l)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không

```python
n=int(input())
x=1
while x<n:
    x=x*3
if x==n:
    print("Day la so dang 3^k")
else:
    print("Day ko phai la so dang 3^k")
```

```python
Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A
```

```python
a=int(input())
n1=1
n2=1
ket_qua=0
for i in range(a):
    n3=n2+n1
    n1=n2
    n2=n3
    if n2<a:
        ket_qua=n2
print(ket_qua)
```

```python
#Bai 4
h=int(input("Nhập chiều cao tam giác:"))
for i in range(1,h+1):
    space = " " * (h-i)
    if i==1:
        print(space + "*")
    elif i==h:
        print("*" * (2*h-1))
    else:
        khoang_trong = " " * (2*i-3)
        print(space + "*" + khoang_trong + "*")

```

```python
#bai 6

#Goi bien:
a, b, c = map(float, input("Nhap 3 canh cua tam giac: ").split())
p= (a +b + c)/2
s= (p*(p-a)*(p-b)*(p-c))**0.5
#Dieu kien
if a + b > c and a + c > b and b + c > a:
    print(f"Dien tich cua tam giac la: {s:.2f}")
else:
    print("Day khong phai la 3 canh cua mot tam giac")
```

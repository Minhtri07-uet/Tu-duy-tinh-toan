```python
#Bai 5
#ax^2+bx+c=0
#-> delta= b^2-4ac
#-> x1=(-b+delta^0.5)/2a x2=(-b-delta^0.5)/2a
#-> delta<0 delta=0 pt có nghiệm kép x=-b/2a
 
#Goi bien

a, b, c= map(float, input("Nhập hệ số của phương trình:").split())
if a!=0:
    delta= (b**2)-4*a*c
    if delta>0:
        x1= (-b-delta**0.5)/(2*a)
        x2= (-b+delta**0.5)/(2*a)
        print("Phương trình có 2 nghiệm phân biệt:","x1 =",x1,"x2 =",x2)
    elif delta==0:
         x=-b/(2*a)
         print("Phương trình có nghiệm kép:","x =",x)
    else: 
         print("Phương trình vô nghiệm")
else: 
    if b==0 and c==0: 
        print("Phương trình có vô số nghiệm:")
    elif b==0 and c!=0:
        print("Phương trình vô nghiệm")
    else: 
        x3=-c/b
        print("Phương trình có nghiệm duy nhất là:","x =",x3)

    
 
```

```python
#Bai 8

#Phuong trinh:
# ax + by =m
# cx + dy =n
# Gọi Biến
a, b, c ,d = map(float, input("Nhập hệ số a,b,c,d:").split())
a1, a2 = map(float, input("Nhập hệ số a1,a2:").split())
det = a*d-b*c
detx = a1*d-a2*b
dety = a*a2-c*a1
if det != 0:
    x = detx/det
    y = dety/det
    print("Hệ có nghiệm duy nhất:","x=",x,"y=",y)
else:
    if detx==0 and dety==0:
     print("Hệ có vô sô nghiệm")
    else:
     print("Hệ vô nghiệm")
    
```

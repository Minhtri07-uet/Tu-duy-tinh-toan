#### W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def shh(a:int):
    if a<6:
        return False
    total=0
    for i in range (1,(a//2)+1):
        if a%i==0:
            total+=i
    return total==a
a=int(input())
if shh(a):
    print("Day la so hoan hao")
else:
    print("Day ko phai la so hoan hao")
```

#### W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử

```python
def tvt(ds:list , k:int):
    for i in range (len(ds)):
        if ds[i]==k:
            return i
    return -1

ds=list(range(1,31))
k=int(input())
print(tvt(ds,k))
```

###### Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
###### Đầu ra: Kết quả phép toán.

#### W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a:int,b:str,op:str):
    result=0
    if op=="+":
        result=a+b
    elif op=="-":
        result=a-b
    elif op=="/":
        if b==0:
            return False
        else:
            result=a/b
    elif op=="*":
        result=a*b
    return result
a,b=map(int,input().split())
op=input()
print(compute(a,b,op))
    

      
         
    
```

#### W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triagle(a:int,b:int,c:int):
    if a+b>c and a+c>b and b+c>a:
        if (a**2)+(b**2)==(c**2) or (a**2)+(c**2)==(b**2) or (c**2)+(b**2)==(a**2):
            return True
        else:
            return False
    else:
        return False
a,b,c=map(int,input().split())
print(is_right_triagle(a,b,c))
```
